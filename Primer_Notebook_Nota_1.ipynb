# 1. Problema sencillo de ingeniería

Supongamos un problema típico en **ingeniería civil o agrícola**:

**Calcular el área de un terreno rectangular.**

Este cálculo es común para:

* estimar área de cultivo
* calcular área de una losa
* dimensionar un terreno

La fórmula es:

[
Area = Largo \times Ancho
]

---

# 2. Definición del algoritmo

Primero definimos el **modelo del problema**.

### Entradas

Datos que el usuario debe ingresar.

* `largo` del terreno (m)
* `ancho` del terreno (m)

### Proceso

Operación matemática:

```
area = largo × ancho
```

### Salida

Resultado que mostrará el programa.

* Área del terreno en **metros cuadrados**

---

# 3. Pseudocódigo del algoritmo

Antes de programar, escribimos la lógica.

```
Inicio

Leer largo
Leer ancho

Calcular area = largo * ancho

Mostrar area

Fin
```

Este paso es clave para cumplir el criterio de **comprensión algorítmica**, que en la rúbrica se evalúa como la capacidad de **plantear un algoritmo claro para resolver el problema**.

---

# 4. Implementación en Python

Ahora llevamos el algoritmo al código.

```python
# Programa para calcular el área de un terreno rectangular
# Aplicación básica en ingeniería

# ENTRADAS
largo = float(input("Ingrese el largo del terreno (m): "))
ancho = float(input("Ingrese el ancho del terreno (m): "))

# PROCESO
area = largo * ancho

# SALIDA
print("El área del terreno es:", area, "metros cuadrados")
```

---

# 5. Ejemplo de ejecución

Si el usuario ingresa:

```
largo = 50
ancho = 20
```

El programa mostrará:

```
El área del terreno es: 1000 metros cuadrados
```
